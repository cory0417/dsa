```python
import matplotlib.pyplot as plt
import pandas as pd
df_raw = pd.read_csv("multiplication_benchmark.csv", sep=", ", header=0)
dimensions = df_raw["dim"]
times_standard = df_raw["standard"]
times_strassen = df_raw["strassen"]

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(dimensions, times_standard, color='blue', alpha=0.5, label='Standard Multiplication')
plt.scatter(dimensions, times_strassen, color='red', alpha=0.5, label='Strassen Multiplication')

plt.title('Benchmark Data: Time(ns) vs Dimensions')
plt.xlabel('Dimensions')
plt.ylabel('Time (ns)')
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.savefig("multiplication_benchmark.pdf")
plt.show()
```
